```python
from astroquery.mast import Catalogs
```

```python
from astropy import units as u

RADIUS = 1 * u.arcsec

def get_data():
    res = Catalogs.query_object(f'HIP 33094', radius=RADIUS, catalog='TIC', pagesize=1, nr=1)
    row = res[0]
    fields = row.dtype.names
    return {f: res[0][f] for f in fields}
```

```python
res = get_data()
```

```python
import numpy as np
np.save("TIC.npy", res)
```
